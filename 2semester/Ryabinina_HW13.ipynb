## Решение

1. Функция $f : X \subset \mathbb{R}^n \to \mathbb{R}$ называется сильно выпуклой с константой $m > 0$, если $X$ - выпуклое множество  и $\forall x_1, x_2 \in X$ и $ \alpha \in [0,1]$ выполнено: $$f(\alpha x_1 + (1-\alpha) x_2) \leqslant \alpha f(x_1) + (1-\alpha) f(x_2) - \frac{m}{2} \alpha(1-\alpha)\|x_1 - x_2\|_2^2$$ Покажем сильную выпуклость $\|x\|_2^2$ по определению: $$\alpha^2\|x_1\|_2^2 + (1-\alpha)^2\|x_2\|_2^2 + 2\alpha(1-\alpha) (x_1,x_2) \leqslant \alpha\|x_1\|_2^2 + (1-\alpha)\|x_2\|_2^2 -\frac{m}{2}\alpha(1-\alpha) (\|x_1\|_2^2 + \|x_2\|_2^2 - 2(x_1,x_2))$$ Положим $m = 2$, это и будет константа сильной выпуклости. Тогда получим $$\alpha^2\|x_1\|_2^2 + (1-\alpha)^2\|x_2\|_2^2 + 2\alpha(1-\alpha) (x_1,x_2) \leqslant (\alpha - \alpha + \alpha^2) \|x_1\|_2^2 + (1-\alpha  - \alpha + \alpha^2)\|x_2\|_2^2 + \alpha(1-\alpha)2(x_1,x_2)$$ А это является тождеством $\forall x_1, x_2 \in \mathbb{R}^n, \forall \alpha \in [0,1]$.

2. Пусть $ A = (x_1, f(x_1)), B = (x_2, f(x_2))$,  уравнение:
$$\alpha f(x_1) + (1-\alpha) f(x_2)$$ задает уравнение секущей, проходящей через точки A и B. Уравнение: $$\alpha f(x_1) + (1-\alpha) f(x_2) - \frac{m}{2} \alpha(1-\alpha)\|x_1 - x_2\|_2^2 $$ задает уравнение параболы, проходящей через точки A и B. Получаем, что неравенство из определения задает условие на то, что график функции расположен ниже параболы (и секущей). При этом прогиб параболы опеределяется константой $m$. Иначе можно сказать, что в области между секущей и функцией можно построить параболу.

3. Имеем $\|f'(x) - f'(y)\|_2 \leqslant L\|x - y\|$, покажем что $\|f''(x)\|_2 \leqslant L$. По опеределению: 

$$ f''(x) = \lim_{y \to x} \frac{f'(x) - f'(y)}{x-y}$$ тогда $$\|f''(x)\|_2 = \|\lim_{y \to x} \frac{f'(x) - f'(y)}{x-y}\|_2$$ Норма $\|  \|_2$ непрерывна в т. $x_0 = \frac{f'(x) - f'(y)}{x-y}$, тогда по теореме из курса мат. анализа получаем $$ \|\lim_{y \to x} \frac{f'(x) - f'(y)}{x-y}\|_2  = \lim_{y \to x} \|\frac{f'(x) - f'(y)}{x-y}\|_2$$ Откуда получаем, что $\|f''(x)\|_2 \leqslant L$. 

Операторная норма: $$\|A\| =\sup\{\|Ax\|:x\in \mathbb{R}^n,\ \|x\|=1\} =\sup \left\{{\frac {\|Ax\|}{\|x\|}}:x\in \mathbb{R}^n,\ x\neq 0\right\}$$

Oператорная норма, порождённая второй векторной нормой. Покажем, что если $(Ax,y) = (x, Ay)$, то   

$$\|A\|_2 =\max \left\{{\frac {\|Ax\|}{\|x\|}}:x\in \mathbb{R}^n,\ x\neq 0\right\} = \max \{ \lambda_j: j = 1,2\dots, n; \lambda_j - \text{собственные числа оператора } A\}$$

В случае $A = A^*$ существует ортонормированный базис $e_1, e_2 \dots, e_n$, состоящий из собственных векторов  $A$: 
$$Ae_j = \lambda_je_j, j = 1,2\dots, n$$
Рассмотрим $x \in \mathbb{R}^n$: 
$$x = x_1e_1+ \dots +x_ne_n$$ $$Ax = \lambda_1x_1e_1 + \dots + \lambda_nx_ne_n$$
Так как базис ортонормированный 
$$\|x\| = (x_1^2 + \dots + x_n^2)^{\frac{1}{2}}$$  $$\|Ax\| = ((\lambda_1x_1)^2 + \dots + (\lambda_nx_n)^2)^{\frac{1}{2}}$$
Получаем 
$$ \|Ax\| \leqslant \max\{|\lambda_j|(x_1^2 + \dots + x_n^2)^{\frac{1}{2}}: j = 1,2\dots, n\} = \max |\lambda_j|\|x\|$$
Для любого $x \in \mathbb{R}^n$ выполнено $\frac{\|Ax\|}{\|x\|} \leqslant \max |\lambda_j|$, а для $x = e_k$, $\max|\lambda_j| = |\lambda_k| $ выполняется $\frac{\|Ae_k\|}{\|e_k\|} = |\lambda_k| = \max |\lambda_j|$. Следовательно, доказываемое равенство выполняется.

````python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

f = lambda x: np.exp(x[0] + 3 * x[1] - 0.1) + np.exp(-x[0] - 0.1) + np.exp(x[0] - 3 * x[1] - 0.1)
grad_f = lambda x: np.array([np.exp(x[0] + 3 * x[1] - 0.1) - np.exp(-x[0] - 0.1) + np.exp(x[0] - 3 * x[1] - 0.1),
                           3 * np.exp(x[0] + 3 * x[1] - 0.1) - 3 * np.exp(x[0] - 3 * x[1] - 0.1)])
x0 = np.array([6,7])

def backtracking(x, descent_dir, **kwargs):
    f = kwargs["f"]
    grad_f = kwargs["grad_f"] 
    if kwargs["method"] == "Armijo":
        beta1 = kwargs["beta1"]
        rho = kwargs["rho"]
        alpha = 1
        while f(x + alpha * descent_dir) >= f(x) + beta1 * alpha * grad_f(x).dot(descent_dir) or \
                 np.isnan(f(x + alpha * descent_dir)):
            alpha *= rho
            if alpha < 1e-16:
                break
    return alpha

def GradientDescent_P(f, gradf, x0, epsilon, num_iter, line_search, P, 
                    disp=False, callback=None, **kwargs):
    x = x0.copy()
    iteration = 0
    opt_arg = {"f": f, "grad_f": gradf}
    for key in kwargs:
        opt_arg[key] = kwargs[key]
    while True:
        gradient = P.dot(gradf(np.linalg.inv(P).dot(x))) 
        alpha = line_search(np.linalg.inv(P).dot(x), -gradient, **opt_arg)
        x = x - alpha * gradient
        if callback is not None:
            callback(np.linalg.inv(P).dot(x))
        iteration += 1
        if disp:
            print("Current function val =", f(x))
            print("Current gradient norm = ", np.linalg.norm(gradf(x)))
        if np.linalg.norm(np.linalg.inv(P).dot(gradient)) < epsilon:
            break
        if iteration >= num_iter:
            break
    res = {"x": x, "num_iter": iteration, "tol": np.linalg.norm(gradf(x))}
    return res

def my_callback(x, array):
    array.append(x)

def result(P):
    x_conv = []

    callback = lambda x: my_callback(x, x_conv)

    res = GradientDescent_P(f, grad_f, x0, 1e-4, 100, backtracking, P, callback=callback,
                            method="Armijo", beta1=0.3, rho=0.8)
    grad_conv = [np.linalg.norm(grad_f(x)) for x in x_conv]
    plt.figure(figsize=(8,6))
    plt.semilogy(grad_conv, label=r"$\| f'(x_k) \|_2$")

    plt.semilogy([np.linalg.norm(np.array(x_conv[i]) - np.array(x_conv[i+1])) for i in range(len(x_conv) - 1)], label=r"$\|x_k - x_{k+1}\|_2$")
    plt.semilogy([np.linalg.norm(f(np.array(x_conv[i])) - f(np.array(x_conv[i+1]))) for i in range(len(x_conv) - 1)], label=r"$\|f(x_k) - f(x_{k+1})\|_2$")
    plt.xlabel(r"Number of iteration, $k$", fontsize=20)
    plt.ylabel(r"Convergence rate", fontsize=20)
    plt.xticks(fontsize = 20)
    plt.yticks(fontsize = 20)
    plt.legend(loc="best", fontsize=20)

    print("Number of iteration = {}".format(res["num_iter"]))
    print("Min_x = {}".format(np.linalg.inv(P).dot(res['x'])))
````

### Решение

Минимум функции в точке $x_{min} = (1,1)$,  $f(x_{min}) = 0$.

````python
def GradientDescent(f, gradf, x0, epsilon, num_iter, line_search, 
                    disp=False, callback=None, callback_a=None, **kwargs):
    x = x0.copy()
    iteration = 0
    opt_arg = {"f": f, "grad_f": gradf}
    for key in kwargs:
        opt_arg[key] = kwargs[key]
    while True:
        gradient = gradf(x)
        alpha = line_search(x, -gradient, **opt_arg)
        x = x - alpha * gradient
        if callback is not None:
            callback(x)
        if callback_a is not None:
            callback_a(np.linalg.norm(alpha * gradient))
        iteration += 1
        if disp:
            print("Current function val =", f(x))
            print("Current gradient norm = ", np.linalg.norm(gradf(x)))
        if np.linalg.norm(gradf(x)) < epsilon:
            break
        if iteration >= num_iter:
            break
    res = {"x": x, "num_iter": iteration, "tol": np.linalg.norm(gradf(x))}
    return res
````

````python
f_r = lambda x: 100 * (x[1] - x[0]**2)**2 + (1 - x[0])**2
grad_f_r = lambda x: np.array([100 * (-4 * x[1] * x[0] + 4 * x[0]**3) - 2 + 2 * x[0], 100 * (2 * x[1] - 2 * x[0]**2)])
````

````python
alphas = []
callback_a = lambda x: my_callback(x, alphas)

x0 = [1.2, 1.2]
res = GradientDescent(f_r, grad_f_r, x0, 1e-4, 100, backtracking, callback_a= callback_a,
                            method="Armijo", beta1=0.3, rho=0.8)
res
````
